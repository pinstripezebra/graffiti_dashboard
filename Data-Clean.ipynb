```python
import pandas as pd
```

### Loading Tabular Data

```python
# Census Data
census_df = pd.read_csv('./Data/census-data.csv')
census_df.columns = census_df.iloc[0]
census_df = census_df.drop(census_df.index[0])

numeric_cols = ['Estimate!!Households!!Total!!Less than $10,000',
                             'Estimate!!Households!!Total!!$10,000 to $14,999',
                             'Estimate!!Households!!Total!!$15,000 to $24,999',
                             'Estimate!!Households!!Total!!$25,000 to $34,999',
                             'Estimate!!Households!!Total!!$35,000 to $49,999',
                             'Estimate!!Households!!Total!!$50,000 to $74,999',
                             'Estimate!!Households!!Total!!$75,000 to $99,999',
                             'Estimate!!Households!!Total!!$100,000 to $149,999',
                             'Estimate!!Households!!Total!!$150,000 to $199,999',
                             'Estimate!!Households!!Total!!$200,000 or more',
                             'Estimate!!Households!!Median income (dollars)']
# Taking subset of columns
census_df_clean = census_df[['Geographic Area Name'] + numeric_cols]

# Cleaning subset
census_df_clean[['City', 'State']] = census_df_clean['Geographic Area Name'].str.split(",",expand=True)
census_df_clean['City'] = census_df_clean['City'].map(lambda x: x.rstrip(' city'))
census_df_clean[numeric_cols] = census_df_clean[numeric_cols].apply(pd.to_numeric, errors='coerce')
census_df_clean.head(10)
```

```python
# Top 50 Cities
city_df = pd.read_csv('./Data/CityList.csv')
print(city_df.columns)
city_df.head(10)
```

### Merging DataSets

```python
result = pd.merge(city_df, census_df_clean, how="left", on=["City", "City"])
result.dropna()
result = result[result['Estimate!!Households!!Median income (dollars)'] > 0]
result.head(10)
```

### Adding Graffiti DataSet

```python
graffiti_df = pd.read_csv('./Data/Image Count by City.csv')
graffiti_df.head(10)
```

```python
combined = pd.merge(result, graffiti_df, how="left", on=["City", "City"])
combined = combined.dropna()
combined.head(10)
```

### Writing Final Output

```python
combined.to_csv('./Data/combined.csv')
```
